# Descriptive Statistics
- It uses data that describes the population either through numerical calculations,graphs or tables.It provides a graphical summary of data.

# Inferential Statistics
- It make inferences and predictions about the population based on a sample of data taken from the population.It generalizes a large dataset and applies probabilities to conclude.

# Statistic Vs Parameter

- Statistic is a characterist of a sample.The goal of statistic inference is to use the information obtained from the sample to make inferences about the population parameters.

- Parameter is a characteristic of a population.parameter are generally unknown and are estimated using statistics.


# Topics Under Inferential Statistics
- 1. Population and Sample

        - Population: The complete set of all individuals or items of interest.

        - Sample: A smaller subset taken from the population to draw conclusions.

        - Samples are used because studying the whole population is often impractical.

- 2. Sampling Methods

        - Techniques used to select a representative sample.

        - Random Sampling: Equal chance for every member.

        - Stratified Sampling: Population divided into strata; samples taken from each.

        - Systematic Sampling: Every k-th element is selected.

        - Cluster Sampling: Entire clusters are randomly selected.

- 3. Estimation

        - Estimation is the process of using sample data to estimate population parameters.

        - Examples: population mean, proportion, variance

- 4. Point Estimation

        - A single numerical value used to estimate a population parameter.

        - Sample mean (x)  estimates population mean ()

        - Sample proportion (p)  estimates population proportion (p)

- 5. Confidence Intervals

        - A range of values within which the population parameter is likely to fall.

        - Usually expressed with confidence levels like 90%, 95%, or 99%

        - Wider intervals mean more confidence

- 6. Hypothesis Testing

        - A formal method to test assumptions about a population.

        - Null Hypothesis (H): No effect or difference

        - Alternative Hypothesis (H): There is an effect or difference

- 7. p-value and Significance Level ()

        - p-value: Probability of observing the result assuming H is true

        - Significance level (): Threshold for rejecting H (commonly 0.05)

        - If p-value    Reject H

- 8. Z-test

        - Used to test hypotheses when:

        - Sample size is large (n  30)

        - Population variance is known

        - Commonly used for means and proportions.

- 9. t-test

        - Used when:

        - Sample size is small (n < 30)

        - Population variance is unknown

        - Types:

        - One-sample t-test

        - Two-sample t-test

        - Paired t-test

- 10. Chi-square Test

        - Used for categorical data.

        - Tests association between variables

        - Tests goodness of fit

- 11. ANOVA (Analysis of Variance)

        - Used to compare means of three or more groups.

        - Determines whether at least one group mean differs

- 12. Correlation and Regression
        - Correlation

        - Measures strength and direction of relationship between variables.

        - Values range from -1 to +1

        - Regression

        - Models the relationship between dependent and independent variables.

        - Used for prediction and trend analysis

- 13. Standard Error

        - Measures the variability of a sample statistic from the population parameter.

        - Indicates accuracy of an estimate

        - Smaller standard error  more reliable estimate

        - Formula:

        - Standard Error =  / n

        - where:

        -  = population standard deviation

        - n = sample size

# Measures of Central Tendency

## 1. Central Tendency (Definition)

Central tendency refers to statistical measures that **describe the center or typical value** of a dataset. It helps summarize data using a single representative value.

---

## 2. Why Central Tendency is Important

* Simplifies large datasets
* Helps compare different datasets
* Provides a quick overall idea of data distribution

---

## 3. Main Measures of Central Tendency

### 3.1 Mean (Arithmetic Mean)

The **average** of all values in a dataset.

**Formulas:**

* **Population Mean ():**  = (X) / N
* **Sample Mean (x):** x = (x) / n

where:

* X / x = sum of all observations
* N = population size, n = sample size

**Advantages:**

* Uses all data values
* Easy to calculate

**Disadvantages:**

* Highly affected by extreme values (outliers)

---

### 3.2 Weighted Mean

The **mean where each observation has a different weight** or importance.

**Formula:**
Weighted Mean = ((wi * xi)) / wi

where:

* xi = observation
* wi = weight of observation

**Use:**

* Useful when some data points contribute more than others.

---

### 3.3 Trimmed Mean

The **mean after removing a certain percentage of extreme values** from both ends of the dataset.

**Use:**

* Reduces the effect of outliers
* Often used in highly skewed data

**Example:**

* 10% trimmed mean  remove top 10% and bottom 10% of values before calculating mean

---

### 3.4 Median

The **middle value** when data is arranged in ascending or descending order.

**Steps to find Median:**

* Arrange data in order
* If n is odd  middle value
* If n is even  average of two middle values

**Advantages:**

* Not affected by outliers
* Best for skewed data

**Disadvantages:**

* Does not use all observations

---

### 3.5 Mode

The value that **occurs most frequently** in a dataset.

**Types of Mode:**

* **Unimodal:** One mode
* **Bimodal:** Two modes
* **Multimodal:** More than two modes
* **No Mode:** All values occur equally

**Advantages:**

* Can be used with categorical data
* Easy to identify

**Disadvantages:**

* May not represent the dataset well

---

## 4. Comparison of Mean, Median, and Mode

| Measure       | Definition            | Affected by Outliers | Best Use                                    |
| ------------- | --------------------- | -------------------- | ------------------------------------------- |
| Mean          | Average value         | Yes                  | Symmetric data                              |
| Weighted Mean | Weighted average      | Yes (weights matter) | When observations have different importance |
| Trimmed Mean  | Mean without extremes | Less affected        | Skewed data                                 |
| Median        | Middle value          | No                   | Skewed data                                 |
| Mode          | Most frequent value   | No                   | Categorical data                            |

---

## 5. Relationship Between Mean, Median, and Mode

* **Symmetric Distribution:** Mean = Median = Mode
* **Positively Skewed:** Mean > Median > Mode
* **Negatively Skewed:** Mean < Median < Mode

---

## 6. Conclusion

Measures of central tendency provide a **simple and effective summary of data**, helping in analysis, comparison, and decision-making. Mean, weighted mean, trimmed mean, median, and mode each have their own importance depending on the nature of data.


# Measures of Variability

## 1. Variability (Definition)

Variability (or dispersion) refers to the extent to which data points in a dataset **differ from each other and from the central value**. It shows how spread out the data is.

---

## 2. Importance of Measures of Variability

* Helps understand the spread of data
* Complements measures of central tendency
* Identifies consistency or volatility

---

## 3. Main Measures of Variability

### 3.1 Range

The **difference between the maximum and minimum values** in the dataset.

**Formula:**
Range = Maximum value - Minimum value

**Advantages:**

* Simple to calculate
* Provides quick idea of spread

**Disadvantages:**

* Highly affected by extreme values
* Ignores the distribution of other data points

---

### 3.2 Interquartile Range (IQR)

The **range of the middle 50% of data**, calculated as Q3 - Q1.

* **Q1 (First Quartile):** 25th percentile
* **Q3 (Third Quartile):** 75th percentile

**Formula:**
IQR = Q3 - Q1

**Advantages:**

* Not affected by outliers
* Shows spread of central data

---

### 3.3 Variance

Measures the **average squared deviation from the mean**.

**Formulas:**

* **Population Variance ():**  = (X - ) / N
* **Sample Variance (s):** s = (x - x) / (n - 1)

**Advantages:**

* Uses all data points
* Basis for standard deviation and other statistical analysis

**Disadvantages:**

* In squared units (not same as original data units)

---

### 3.4 Standard Deviation

The **square root of variance**, measures dispersion in same units as data.

**Formulas:**

* **Population SD ():**  = ((X - ) / N)
* **Sample SD (s):** s = ((x - x) / (n - 1))

**Advantages:**

* Most widely used measure of spread
* Same units as original data

**Disadvantages:**

* Sensitive to outliers

---

### 3.5 Coefficient of Variation (CV)

Measures **relative variability**, useful for comparing datasets with different units or means.

**Formula:**
CV = (Standard Deviation / Mean)  100%

**Advantages:**

* Unitless
* Allows comparison across datasets

**Disadvantages:**

* Misleading if mean  0

---

### 3.6 Mean Absolute Deviation (MAD)

Average of **absolute deviations from the mean**.

**Formula:**
MAD = |x - x| / n

**Advantages:**

* Easy to understand
* Less sensitive to extreme values than variance

**Disadvantages:**

* Less commonly used

---

## 4. Summary Table

| Measure                       | Formula                     | Advantage                                | Disadvantage                             |
| ----------------------------- | --------------------------- | ---------------------------------------- | ---------------------------------------- |
| Range                         | Max - Min                   | Simple, quick                             | Affected by outliers, ignores other data |
| IQR                           | Q3 - Q1                     | Resistant to outliers                     | Does not consider extremes               |
| Variance                      | (X-)/N or (x-x)/(n-1)| Uses all data, foundation for SD          | In squared units                          |
| Standard Deviation             | Variance                   | Same units as data, widely used           | Sensitive to outliers                     |
| Coefficient of Variation (CV) | (SD / Mean)  100%          | Unitless, allows comparison               | Misleading if mean  0                    |
| Mean Absolute Deviation (MAD) | lx - xl / n               | Easy to interpret, less affected by outliers | Less common                             |


## 5. Conclusion

Measures of variability provide **important insight into the spread and consistency of data**. They complement measures of central tendency to fully describe a dataset.

# Graphical Representation in Statistics and Univariate Analysis

## 1. Introduction

Graphical representation is a **visual way to present data**, helping to understand patterns, trends, and distributions quickly. In **univariate analysis**, we analyze one variable at a time, and graphs help visualize the distribution and spread of that variable.

---

## 2. Common Graphs

### 2.1 Histogram

* Represents **frequency distribution** of a continuous variable.
* Data divided into **intervals (bins)**.
* Identifies distribution shape (normal, skewed, etc.).

**Example:** Distribution of student scores.

### 2.2 Bar Chart

* Represents **categorical data**.
* Height of bars corresponds to frequency.

**Example:** Number of students in different departments.

### 2.3 Pie Chart

* Represents **proportions of categories**.
* Each sector shows percentage of total.

**Example:** Market share of smartphone brands.

### 2.4 Box Plot (Whisker Plot)

* Shows **spread, median, quartiles, and outliers**.
* Useful for detecting variability.

**Example:** Salary distribution.

### 2.5 Frequency Polygon

* Line graph connecting **midpoints of class intervals**.
* Shows trend of distribution.

**Example:** Distribution of temperatures over a month.

### 2.6 Line Graph

* Shows **trend over time**.
* Points connected by line.

**Example:** Temperature changes over a week.

### 2.7 Scatter Plot

* Shows **relationship between two numerical variables**.
* Each point represents a data pair (x, y).

**Example:** Height vs Weight.

### 2.8 Stem-and-Leaf Plot

* Displays **distribution while preserving individual data points**.
* Useful for small datasets.

**Example:** Exam scores of a small class.

---

## 3. Summary Table

| Graph Type        | Data Type           | Purpose                                 |
| ----------------- | ------------------- | --------------------------------------- |
| Histogram         | Continuous          | Show frequency distribution             |
| Bar Chart         | Categorical         | Compare categories                      |
| Pie Chart         | Categorical         | Show proportions                        |
| Box Plot          | Continuous          | Display spread, median, outliers        |
| Frequency Polygon | Continuous          | Show distribution trends                |
| Line Graph        | Continuous/Time     | Observing trends over time              |
| Scatter Plot      | Numerical           | Examine relationships between variables |
| Stem-and-Leaf     | Continuous/Discrete | Preserve individual data points         |

---

## 4. Conclusion

Graphical representations, especially for univariate analysis, **simplify data interpretation**, reveal patterns, trends, and distributions, and make statistical analysis more intuitive and effective.
