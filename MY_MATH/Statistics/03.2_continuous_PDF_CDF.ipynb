# ðŸ“Š Probability Density Function (PDF)

A **Probability Density Function (PDF)** describes the probability distribution of a **continuous random variable** ($X$). It shows how densely probability is distributed over the values of the variable.

---

## ðŸ“œ Definition

Let $X$ be a continuous random variable. If the probability that $X$ falls within a small interval $dx$ around $x$ is given by:

$$
P(x < X < x + dx) = f(x)\,dx
$$

then $f(x)$ is called the **Probability Density Function (PDF)** of $X$.

---

## âœ… Properties of PDF

1.  **Non-negativity:**
    $$f(x) \ge 0 \quad \text{for all } x$$

2.  **Total Probability (Normalization):** The total area under the PDF curve must equal 1.
    $$
    \int_{-\infty}^{\infty} f(x)\,dx = 1
    $$

3.  **Probability over an Interval:** The probability that $X$ falls between $a$ and $b$ is the area under the curve from $a$ to $b$.
    $$
    P(a \le X \le b) = \int_a^b f(x)\,dx
    $$

4.  **Probability at a Single Point:**
    $$
    P(X = x) = 0
    $$

---

## âœï¸ Example Calculation

Given the PDF:
$$
f(x) = kx^3, \quad 0 \le x \le 3
$$

### 1. Finding the Constant $k$

$$
\int_0^3 kx^3 dx = 1
$$
$$
k \left[ \frac{x^4}{4} \right]_0^3 = 1
$$
$$
k \left( \frac{81}{4} \right) = 1
$$
$$
\mathbf{k = \frac{4}{81}}
$$

### 2. Finding the Probability $P(1 < X < 2)$

$$
P(1 < X < 2) = \int_1^2 \frac{4}{81}x^3 dx
$$
$$
P = \frac{4}{81} \left[ \frac{x^4}{4} \right]_1^2
$$
$$
P = \frac{4}{81} \cdot \frac{15}{4}
$$
$$
\mathbf{P(1 < X < 2) = \frac{15}{81} = \frac{5}{27}}
$$

---

## ðŸ“ˆ Key Statistical Measures

### Mean (Expected Value)
$$
E(X) = \mu = \int_{-\infty}^{\infty} x f(x)\,dx
$$

### Variance
$$
Var(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x)\,dx
$$

# ðŸ–¼ï¸ The Meaning of Area in a Probability Density Function (PDF)

In a **PDF**, the y-axis represents **Probability Density**, not probability itself. To find actual probability, we must calculate the **Area** under the curve.

---

### 1. The Core Rules
1. **The Height is not Probability:** While the y-axis shows "density," the height at a single point does not give you the chance of that value occurring.
2. **Total Area = 1:** The area under the entire curve must always equal **1**, representing a 100% total probability.
3. **Partial Area = Specific Probability:** The area between two points $a$ and $b$ represents the probability that the random variable $X$ falls within that interval: 
   $$P(a \le X \le b)$$



---

### 2. Why do we use Area?
In a continuous distribution (like measuring time, weight, or height), the probability of hitting an **exact** single point (e.g., exactly $1.750000...$ meters) is technically **zero**. 

Because we cannot measure points, we measure **"slices"** (intervals). 
* A **wider slice** contains more area and represents a **higher probability**.
* A **narrower slice** contains less area and represents a **lower probability**.

---

### 3. Practical Example: Uniform Distribution
Imagine a random variable $X$ that can be any value between **0 and 4**.

* **The Width:** 4 units.
* **The Height (Density):** To ensure the total area is 1, the height must be $1/4 = 0.25$.
  * $\text{Total Area} = 4 \times 0.25 = 1$.



**Scenario:** What is the probability that $X$ is between **1 and 3**?

1. **Calculate the Width:** $3 - 1 = 2$ units.
2. **Calculate the Area:** $\text{Width} \times \text{Height} = 2 \times 0.25 = 0.5$.
3. **Result:** There is a **50% probability** ($0.5$) that the value falls between 1 and 3.

---

**Summary for Analysis:** In a PDF, **height** tells you the relative likelihood (density), but only the **area** gives you the actual probability.

# ðŸ“ˆ How a Probability Density Function (PDF) Graph is Calculated

A PDF graph is not a plot of raw counts; it is a calculated "density." There are two primary ways to calculate the values for the y-axis: **Empirically** (using your data) or **Theoretically** (using a formula).

---

### 1. Empirical Calculation (Kernel Density Estimation - KDE)
When you ask software (like Seaborn or Scipy) to plot a "Density Plot" for your data, it uses a process called **Kernel Density Estimation**.

1. **Place a Kernel:** The algorithm places a small "kernel" function (usually a tiny Gaussian/Bell curve) over every single data point in the set.
2. **Summation:** It sums the heights of all these individual tiny curves at every point along the x-axis.
3. **Normalization:** The final curve is scaled so that the **total area under the curve equals 1**.



---

### 2. Normalized Histogram Method
A PDF can be thought of as a smoothed, normalized histogram. The height (density) of each bar is calculated as:

$$\text{Density} = \frac{\text{Count in Bin}}{\text{Total Samples} \times \text{Bin Width}}$$

* This calculation ensures that the units on the y-axis represent **Probability per Unit of X**, and that all bars combined have an area of 1.



---

### 3. Theoretical Calculation (Fitting a Model)
If you assume your data follows a specific distribution (like a Normal Distribution), you calculate the graph points using a fixed mathematical formula.

**Step 1:** Calculate the data's **Mean ($\mu$)** and **Standard Deviation ($\sigma$)**.
**Step 2:** Plug every $x$ value into the Probability Density Function formula:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

*The result $f(x)$ provides the exact height of the curve at any point $x$.*

---

### ðŸ“‹ Calculation Summary Table

| Method | When to use it? | How $y$ is determined |
| :--- | :--- | :--- |
| **KDE (Empirical)** | When you have raw data but no known formula. | Summing small "kernels" at each point. |
| **Normalized Hist** | Quick data exploration. | Dividing counts by (Total N $\times$ Bin Width). |
| **Theoretical PDF** | When data follows a known shape (e.g., Normal). | Using a fixed math formula ($\mu, \sigma$). |

---

**Key Takeaway:** Regardless of the calculation method, the y-axis is always scaled so that the **integral (total area)** of the graph is exactly **1**.
