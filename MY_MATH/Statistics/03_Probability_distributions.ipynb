#  Random Variables

A **Random Variable (RV)** is a function that assigns a real number to each outcome in the sample space ($S$) of a random experiment. It connects **theoretical probability** with **real-world data**.

**Mathematical Definition:**
$$
X: S \to \mathbb{R}
$$

Where:
* $X$ = The Random Variable
* $S$ = The Sample Space (set of all possible outcomes)
* $\mathbb{R}$ = The Set of Real Numbers (the range of the RV)



---

##  Types of Random Variables

### 1. Discrete Random Variable
* **Definition:** Takes a **finite or countably infinite** set of distinct values.
* **Probability Function:** **Probability Mass Function (PMF)**, $P(X=x_i)$.
    $$
    0 \le P(X=x_i) \le 1, \quad \sum_{i} P(X=x_i) = 1
    $$
* **Example:** Toss two coins, let $X$ = number of heads.

| Sample Space | $X$ (Number of Heads) |
| :---: | :---: |
| HH | 2 |
| HT | 1 |
| TH | 1 |
| TT | 0 |

---

### 2. Continuous Random Variable
* **Definition:** Takes **infinite, uncountable values** within a given range (an interval).
* **Probability Function:** **Probability Density Function (PDF)**, $f(x)$.
    $$
    f(x) \ge 0, \quad \int_{-\infty}^{\infty} f(x)\,dx = 1
    $$
* **Example:** For $f(x) = kx^3$, defined for $0 \le x \le 3$.
    We use the normalization property to find $k$:
    $$
    \int_0^3 kx^3\,dx = 1 \implies k \cdot \frac{x^4}{4}\Big|_0^3 = 1 \implies k \left(\frac{81}{4}\right) = 1 \implies \mathbf{k = \frac{4}{81}}
    $$

---

##  Key Statistical Formulas

### Mean (Expected Value, $E(X)$ or $\mu$)
Represents the long-run average value of the random variable.

* **Discrete:**
    $$
    E(X) = \mu = \sum_{i} x_i \cdot P(X=x_i)
    $$

* **Continuous:**
    $$
    E(X) = \mu = \int_{-\infty}^{\infty} x f(x)\,dx
    $$

### Variance ($Var(X)$ or $\sigma^2$)
Measures the spread or dispersion of the random variable's values around the mean.

The computational formula (applicable to both types) is:
$$
Var(X) = \sigma^2 = E(X^2) - [E(X)]^2
$$

---

#  Probability Distributions

A **probability distribution** describes how probabilities are assigned to all possible outcomes of a **random variable** ($X$). It is the fundamental link between **random variables** and their **probabilities**.



---

##  Definition and Properties

For a random variable $X$, a probability distribution provides a function or table that specifies the likelihood of each outcome.

* **Discrete RV:** Probability Mass Function (PMF): $P(X = x_i)$
* **Continuous RV:** Probability Density Function (PDF): $f(x)$

### Key Properties (Axioms)

1.  **Non-negativity:** All assigned probabilities must be non-negative.
    $$0 \leq P(X=x_i) \leq 1 \quad \text{or} \quad f(x) \ge 0$$

2.  **Normalization:** The total probability over the entire sample space must equal 1.
    * **Discrete:**
        $$\sum_i P(X=x_i) = 1$$
    * **Continuous:**
        $$\int_{-\infty}^{\infty} f(x)\,dx = 1$$

---

##  Types of Distributions

### 1. Discrete Probability Distribution
* **Values:** Takes a finite or countably infinite set of distinct values.
* **Function:** **Probability Mass Function (PMF)** ($P(X = x_i) = p_i$).
* **Example (Rolling a Die):**

| $X$ (Outcome) | 1 | 2 | 3 | 4 | 5 | 6 |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| $P(X)$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ |

---

### 2. Continuous Probability Distribution
* **Values:** Takes infinitely many values within a specified interval.
* **Function:** **Probability Density Function (PDF)** ($f(x)$).
* **Probability Calculation:**
    $$
    P(a \le X \le b) = \int_a^b f(x)\,dx
    $$
* **Example:** For $f(x) = kx^3$, $0 \le x \le 3$.
    $$\int_0^3 kx^3\,dx = 1 \implies \mathbf{k = \frac{4}{81}}$$

---

##  Common Probability Distributions

| Distribution | Type | Mean ($E(X)$) | Variance ($\text{Var}(X)$) | Typical Use Case |
| :---: | :---: | :---: | :---: | :--- |
| **Binomial** | Discrete | $np$ | $np(1-p)$ | Number of successes in $n$ fixed trials. |
| **Poisson** | Discrete | $\lambda$ | $\lambda$ | Number of events in a fixed time/space interval. |
| **Uniform** | Discrete/Continuous | $\frac{a+b}{2}$ | $\frac{(b-a)^2}{12}$ | Equal probability for all outcomes in a range. |
| **Normal (Gaussian)** | Continuous | $\mu$ | $\sigma^2$ | Modeling natural phenomena and measurement errors. |
| **Exponential** | Continuous | $1/\lambda$ | $1/\lambda^2$ | Modeling waiting time until the next event occurs. |

---

##  Key Statistical Formulas

### Expected Value (Mean, $\mu$)
* **Discrete:**
    $$
    E(X) = \sum_i x_i P(X=x_i)
    $$
* **Continuous:**
    $$
    E(X) = \int_{-\infty}^{\infty} x f(x)\,dx
    $$

### Variance ($\text{Var}(X)$)
* **Discrete:**
    $$
    \text{Var}(X) = \sum_i (x_i - \mu)^2 P(X=x_i)
    $$
* **Continuous:**
    $$
    \text{Var}(X) = \int_{-\infty}^{\infty} (x-\mu)^2 f(x)\,dx
    $$

---

##  Importance and Applications

* **Advantages:** Models real-world randomness; enables **prediction, simulations, and risk analysis**.
* **Use Cases:** **Statistics** (Sample data modeling); **Machine Learning** (Probabilistic models); **Finance** (Stock returns, risk management).

---

##  Conclusion

Probability distributions are essential tools in **statistics and data science**. They provide the mathematical framework necessary for analyzing random events, modeling uncertainty, and making informed predictions.

#  Parameters in Probability Distributions

**Parameters** are the fundamental numerical characteristics of a probability distribution that completely define its **shape, location (center), and scale (spread)**. A specific distribution type (e.g., Normal) becomes a specific, usable model when its parameters are assigned definite values.

---

##  Defining Parameters

The most general and universally used parameters are the Mean and Variance, as they describe the center and spread of *any* distribution.

### 1. Central Tendency: Mean (Expected Value, $\mu$)

The weighted average value of a random variable, representing the center of the distribution.

* **Discrete RV:**
    $$
    \mu = E(X) = \sum_i x_i P(X=x_i)
    $$
* **Continuous RV:**
    $$
    \mu = E(X) = \int_{-\infty}^{\infty} x f(x)\,dx
    $$

### 2. Variability: Variance ($\sigma^2$)

Measures the expected squared distance of a random variable from its mean, defining the spread.

* **Discrete RV:**
    $$
    \sigma^2 = \text{Var}(X) = \sum_i (x_i - \mu)^2 P(X=x_i)
    $$
* **Continuous RV:**
    $$
    \sigma^2 = \text{Var}(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x)\,dx
    $$

### 3. Variability: Standard Deviation ($\sigma$)

The square root of variance; it measures spread in the same units as the random variable $X$.
$$
\sigma = \sqrt{\text{Var}(X)}
$$

---

##  Distribution-Specific Parameters

Specific distributions have unique parameters that govern their behavior. 

| Distribution | Notation | Core Parameters | Role in Defining Shape |
| :---: | :---: | :---: | :--- |
| **Binomial** | $X \sim \text{Bin}(n, p)$ | $n$ (trials), $p$ (success probability) | $n$ limits the range; $p$ determines the skewness. |
| **Poisson** | $X \sim \text{Pois}(\lambda)$ | $\lambda$ (rate parameter) | $\lambda$ controls both the mean and the variance ($\mu = \sigma^2 = \lambda$). |
| **Normal** | $X \sim N(\mu, \sigma^2)$ | $\mu$ (mean), $\sigma^2$ (variance) | $\mu$ shifts the curve left/right; $\sigma^2$ makes the curve wider/narrower. |
| **Uniform** | $X \sim U(a, b)$ | $a$ (min), $b$ (max) | $a$ and $b$ define the boundaries and the height of the rectangular PDF. |
| **Exponential** | $X \sim \text{Exp}(\lambda)$ | $\lambda$ (rate parameter) | $\lambda$ controls the decay rate of the probability curve. |

---

##  Importance in Statistical Modeling

1.  **Identity:** Parameters distinguish one instance of a distribution from another (e.g., a Normal distribution with $\mu=5$ is fundamentally different from one with $\mu=10$).
2.  **Estimation:** Statistical inference is largely the process of using sample data to **estimate** the unknown population parameters.
3.  **Simulation & Prediction:** Once parameters are known, the distribution can be used to calculate probabilities for future events and run simulations (e.g., predicting the probability of 5 successes in 10 trials using $\text{Bin}(10, 0.4)$).
4.  **Comparison:** Parameters allow distributions from different experiments to be quantitatively compared.

---

##  Conclusion

Parameters are the **input values** that turn a theoretical probability function into a specific, actionable model. Understanding how $\mu$, $\sigma$, and other specific parameters like $n$ and $p$ define the **shape, location, and spread** is crucial for all statistical analysis and real-world modeling.
