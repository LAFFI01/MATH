# üé≤ Probability Mass Function (PMF)

A **Probability Mass Function (PMF)** describes the probability distribution of a **discrete random variable** ($X$). It assigns a specific probability to **each possible, countable value** of the random variable.



---

## üìú Definition

Let $X$ be a discrete random variable that can take a countable set of values $x_1, x_2, x_3, \dots, x_n$.

The **Probability Mass Function (PMF)** of $X$ is defined as the probability that $X$ takes the value $x_i$:

$$
P(X = x_i) = p(x_i)
$$

where $p(x_i)$ is the probability assigned to the outcome $x_i$.

---

## ‚úÖ PMF Conditions (Crucial Properties)

1.  **Probability Range:** The probability for any single outcome must be between 0 and 1, inclusive.
    $$0 \le p(x_i) \le 1$$

2.  **Total Probability:** The sum of probabilities for all possible outcomes must equal 1.
    $$
    \sum_{i=1}^{n} p(x_i) = 1
    $$

---

## üßÆ How to Calculate Probability Using PMF

Since $X$ is discrete, probabilities are calculated using summation (not integration, as in PDF).

* **Probability of a single value:**
    $$P(X = x) = p(x)$$

* **Probability over a range/set of values:**
    $$P(a \le X \le b) = \sum_{x=a}^{b} p(x)$$

---

## ‚úèÔ∏è Example 1: Tossing a Fair Die

Let $X$ be the outcome when rolling a fair six-sided die.

| $X$ (Outcome $x_i$) | 1 | 2 | 3 | 4 | 5 | 6 |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| $P(X = x_i)$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ | $1/6$ |

**Check:**
$$
\sum_{i=1}^{6} \frac{1}{6} = 1 \quad (\text{Condition met})
$$

---

## üîç Example 2: Finding Missing Probability

Given a PMF table:

| $X$ | 0 | 1 | 2 |
| :---: | :---: | :---: | :---: |
| $P(X)$ | $p$ | $0.3$ | $0.5$ |

Using the Total Probability condition:
$$
p + 0.3 + 0.5 = 1
$$
$$
p = 1 - 0.8
$$
$$
\mathbf{p = 0.2}
$$

---

## üìà Key Statistical Measures

### Mean (Expected Value)
The average value of $X$ over the distribution, denoted by $\mu$.

$$
E(X) = \mu = \sum_{i=1}^{n} x_i \cdot P(X = x_i)
$$

**Example for $p=0.2$ (from Example 2):**
$$
E(X) = 0(0.2) + 1(0.3) + 2(0.5) = 0 + 0.3 + 1.0 = \mathbf{1.3}
$$

### Variance
A measure of the spread or dispersion of the distribution.

$$
Var(X) = \sum_{i=1}^{n} (x_i - \mu)^2 \cdot P(X = x_i)
$$

The calculation is often easier using the computational formula:
$$
Var(X) = E(X^2) - [E(X)]^2
$$

# üìà Cumulative Distribution Function (CDF) of PMF

For a **discrete random variable** ($X$), the **Cumulative Distribution Function (CDF)** is obtained **from the Probability Mass Function (PMF)** by cumulative summation. The CDF captures the probability that the random variable will take a value less than or equal to a specific point $x$.

---

## üìú Definition

Let $X$ be a discrete random variable with PMF given by $P(X = x_i) = p_i$.

The **Cumulative Distribution Function (CDF)** of $X$, denoted $F(x)$, is defined as:

$$
F(x) = P(X \le x)
$$

---

## üßÆ CDF Formula (Discrete Case)

The CDF is calculated by summing the probabilities of all discrete values $x_i$ that are less than or equal to the point $x$:

$$
F(x) = \sum_{x_i \le x} P(X = x_i)
$$

üëâ **Meaning:** $F(x)$ represents the total probability accumulated up to the point $x$.

---

## ‚úÖ Properties of CDF



1.  **Range:** The CDF values are probabilities, so they range from 0 to 1.
    $$0 \le F(x) \le 1$$
2.  **Monotonicity:** $F(x)$ is a **non-decreasing function** (it never goes down).
3.  **Minimum:** As $x$ approaches negative infinity, the accumulated probability is zero.
    $$\lim_{x \to -\infty} F(x) = 0$$
4.  **Maximum:** As $x$ approaches positive infinity, the accumulated probability is one.
    $$\lim_{x \to +\infty} F(x) = 1$$
5.  **Jumps:** The size of the vertical jump in the step function at any point $x_i$ is equal to the probability of that point: $P(X = x_i)$.

---

## ‚úèÔ∏è Example: CDF from PMF

**Given PMF:**

| $X$ | 1 | 2 | 3 | 4 |
| :---: | :---: | :---: | :---: | :---: |
| $P(X)$ | 0.1 | 0.2 | 0.4 | 0.3 |

### Step-by-Step Calculation

$$
F(1) = P(X \le 1) = P(X=1) = \mathbf{0.1}
$$

$$
F(2) = P(X \le 2) = F(1) + P(X=2) = 0.1 + 0.2 = \mathbf{0.3}
$$

$$
F(3) = P(X \le 3) = F(2) + P(X=3) = 0.3 + 0.4 = \mathbf{0.7}
$$

$$
F(4) = P(X \le 4) = F(3) + P(X=4) = 0.7 + 0.3 = \mathbf{1.0}
$$

### CDF Table

| $x$ | $x < 1$ | $1 \le x < 2$ | $2 \le x < 3$ | $3 \le x < 4$ | $x \ge 4$ |
| :---: | :---: | :---: | :---: | :---: | :---: |
| $F(x)$ | 0 | 0.1 | 0.3 | 0.7 | 1.0 |

### Piecewise Representation

The formal representation captures the "step" nature:

$$
F(x) =
\begin{cases}
0, & x < 1 \\
0.1, & 1 \le x < 2 \\
0.3, & 2 \le x < 3 \\
0.7, & 3 \le x < 4 \\
1, & x \ge 4
\end{cases}
$$

---

## üîé How to Find Probability Using CDF

The CDF is most useful for finding probabilities over intervals:

* **Interval Probability:**
    $$
    P(a < X \le b) = F(b) - F(a)
    $$
* **Exact Value (using limits):**
    $$
    P(X = x) = F(x) - \lim_{t \to x^-} F(t)
    $$

---

## üéØ Conclusion

The **CDF of a PMF** is obtained by the **cumulative addition of probabilities**, resulting in a step function that fully describes a discrete random variable. It is the most robust way to calculate probabilities for intervals like $P(X \le x)$ or $P(a < X \le b)$.
