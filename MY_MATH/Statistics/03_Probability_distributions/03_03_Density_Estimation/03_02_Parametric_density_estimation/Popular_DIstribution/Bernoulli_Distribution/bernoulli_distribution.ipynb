# The Bernoulli Distribution: PMF, CDF, and ML Applications

The **Bernoulli Distribution** models a single experiment with exactly two outcomes: **Success (1)** and **Failure (0)**. It is the foundation for binary logic in statistics and machine learning.

---

## 1. Probability Formulas

### Probability Mass Function (PMF)
The PMF gives the probability of a specific outcome $k$:
$$f(k; p) = P(X = k) = p^k (1 - p)^{1 - k} \quad \text{for } k \in \{0, 1\}$$

### Cumulative Distribution Function (CDF)
The CDF, often denoted as $F(x)$, calculates the probability that the random variable $X$ is less than or equal to $x$. Since Bernoulli is discrete, the CDF is a step function:

$$F(x; p) = P(X \le x) = 
\begin{cases} 
0 & x < 0 \\
1 - p & 0 \le x < 1 \\
1 & x \ge 1 
\end{cases}$$

---

## 2. Advanced Use Cases in DS & ML

The Bernoulli distribution is not just for coin flips; it's a core component of many algorithms:

* **Logistic Regression:** Assumes the dependent variable follows a Bernoulli distribution given the input features.
* **Bernoulli Naive Bayes:** A variant of the Naive Bayes classifier used specifically for discrete data where features are binary (e.g., word presence in text classification).
* **Neural Networks (Output Layer):** In binary classification, the final **Sigmoid** activation function outputs a value $\hat{y} \in [0, 1]$, which is treated as the parameter $p$ of a Bernoulli distribution.
* **A/B Testing:** Modeling user conversion (clicked/didn't click) as a Bernoulli trial to calculate statistical significance.

## 3. Python Implementation: PMF & CDF
Use the code below in a **Code Cell** to visualize both the PMF and the "step-like" nature of the CDF.

```python


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Parameters
p_val = 0.4  # Probability of success
x_pmf = [0, 1]
y_pmf = [bernoulli.pmf(k, p_val) for k in x_pmf]

# Continuous range for CDF plot
x_cdf = np.linspace(-0.5, 1.5, 500)
y_cdf = bernoulli.cdf(x_cdf, p_val)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# 1. PMF Plot
ax[0].bar(x_pmf, y_pmf, color=['#ff9999','#66b3ff'], width=0.4)
ax[0].set_xticks(x_pmf)
ax[0].set_xticklabels(['Failure (0)', 'Success (1)'])
ax[0].set_title(f'Bernoulli PMF (p={p_val})')
ax[0].set_ylabel('Probability')

# 2. CDF Plot
ax[1].plot(x_cdf, y_cdf, color='black', drawstyle='steps-post', linewidth=2)
ax[1].fill_between(x_cdf, y_cdf, step="post", alpha=0.1, color='blue')
ax[1].set_title(f'Bernoulli CDF (p={p_val})')
ax[1].set_ylabel('Cumulative Probability')
ax[1].set_ylim(-0.1, 1.1)

plt.tight_layout()
plt.show()
````
