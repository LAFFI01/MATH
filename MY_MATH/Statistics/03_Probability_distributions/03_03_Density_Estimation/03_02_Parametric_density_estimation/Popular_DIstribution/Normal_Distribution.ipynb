# üîî The Normal Distribution (Gaussian Distribution)

The **Normal Distribution** is the most important probability distribution in statistics. It is often called the "Bell Curve" because of its characteristic shape. It describes how the values of a variable are distributed when most observations cluster around a central peak.

---

### 1. Key Characteristics
A perfect Normal Distribution has several unique properties:

* **Symmetry:** The left and right halves are mirror images of each other.
* **Central Tendency:** The **Mean**, **Median**, and **Mode** are all exactly the same and located at the center of the curve.
* **Asymptotic:** The tails of the curve approach the x-axis but never actually touch it (representing that extreme outliers are rare but possible).



---

### 2. The Two Parameters: $\mu$ and $\sigma$
The shape of a Normal Distribution is entirely determined by two numbers:

1.  **Mean ($\mu$):** Determines the **location** of the peak (the center).
2.  **Standard Deviation ($\sigma$):** Determines the **spread** (the width). 
    * A **small $\sigma$** creates a tall, skinny curve.
    * A **large $\sigma$** creates a short, fat curve.

[Image comparing Normal Distributions with different means and standard deviations]

---

### 3. The Empirical Rule (68-95-99.7 Rule)
In a Normal Distribution, data falls into a very predictable pattern regarding the standard deviation:

* **68%** of the data falls within **1** standard deviation ($\mu \pm 1\sigma$).
* **95%** of the data falls within **2** standard deviations ($\mu \pm 2\sigma$).
* **99.7%** of the data falls within **3** standard deviations ($\mu \pm 3\sigma$).



[Image of the 68-95-99.7 rule in a Normal Distribution]


---

### 4. The Probability Density Function (PDF)
The height of the curve at any point $x$ is calculated using this formula:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

* **$\pi$** $\approx 3.14159$
* **$e$** $\approx 2.71828$

---

### 5. Standard Normal Distribution ($Z$-Distribution)
A **Standard Normal Distribution** is a special case where:
* The Mean ($\mu$) = **0**
* The Standard Deviation ($\sigma$) = **1**

We use **$Z$-scores** to "standardize" any normal value into this format, which allows us to compare different datasets:
$$Z = \frac{x - \mu}{\sigma}$$

# üìè The Standard Normal Variate ($Z$)

A **Standard Normal Variate** (often called a **Z-score**) is a specific type of normal random variable that has been "standardized." It represents the number of standard deviations a data point is away from the mean.

---

### 1. Definition
A random variable $X$ is converted into a **Standard Normal Variate ($Z$)** when it is rescaled to have:
* **Mean ($\mu$) = 0**
* **Standard Deviation ($\sigma$) = 1**

Mathematically, we denote this as:  
$$Z \sim N(0, 1)$$



---

### 2. The Standardization Formula (Z-Score)
To transform any value $x$ from a Normal Distribution $N(\mu, \sigma)$ into a Standard Normal Variate $Z$, use the following formula:

$$Z = \frac{x - \mu}{\sigma}$$

* **$x$**: The raw data point.
* **$\mu$**: The mean of the original population.
* **$\sigma$**: The standard deviation of the original population.



---

### 3. Why is $Z$ Important?
Standardizing variables allows us to:
1.  **Compare Apples to Oranges:** You can compare scores from different scales (e.g., comparing an SAT score to an ACT score) by seeing how many standard deviations they are from their respective means.
2.  **Calculate Probabilities Easily:** Instead of calculating complex integrals for every different mean and sigma, we can use a single **Standard Normal Table (Z-Table)**.
3.  **Identify Outliers:** Typically, any $Z$ score greater than $+3$ or less than $-3$ is considered an extreme outlier.

---

### 4. The PDF of the Standard Normal Variate
Because $\mu=0$ and $\sigma=1$, the general Normal PDF simplifies significantly for $Z$:

$$f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$$

---

### üìä Properties of $Z$

| Property | Value |
| :--- | :--- |
| **Mean** | $0$ |
| **Variance** | $1$ |
| **Standard Deviation** | $1$ |
| **Symmetry** | Symmetric around $Z = 0$ |
| **Total Area** | $1.0$ (100% Probability) |

---

### 5. Interpreting $Z$-Scores
* **$Z = 0$**: The value is exactly at the mean.
* **$Z = 1.5$**: The value is 1.5 standard deviations **above** the mean.
* **$Z = -2.0$**: The value is 2.0 standard deviations **below** the mean.
