# üîî The Normal Distribution (Gaussian Distribution)

The **Normal Distribution** is the most important probability distribution in statistics. It is often called the "Bell Curve" because of its characteristic shape. It describes how the values of a variable are distributed when most observations cluster around a central peak.

---

### 1. Key Characteristics
A perfect Normal Distribution has several unique properties:

* **Symmetry:** The left and right halves are mirror images of each other.
* **Central Tendency:** The **Mean**, **Median**, and **Mode** are all exactly the same and located at the center of the curve.
* **Asymptotic:** The tails of the curve approach the x-axis but never actually touch it (representing that extreme outliers are rare but possible).



---

### 2. The Two Parameters: $\mu$ and $\sigma$
The shape of a Normal Distribution is entirely determined by two numbers:

1.  **Mean ($\mu$):** Determines the **location** of the peak (the center).
2.  **Standard Deviation ($\sigma$):** Determines the **spread** (the width). 
    * A **small $\sigma$** creates a tall, skinny curve.
    * A **large $\sigma$** creates a short, fat curve.

[Image comparing Normal Distributions with different means and standard deviations]

---

### 3. The Empirical Rule (68-95-99.7 Rule)
In a Normal Distribution, data falls into a very predictable pattern regarding the standard deviation:

* **68%** of the data falls within **1** standard deviation ($\mu \pm 1\sigma$).
* **95%** of the data falls within **2** standard deviations ($\mu \pm 2\sigma$).
* **99.7%** of the data falls within **3** standard deviations ($\mu \pm 3\sigma$).



[Image of the 68-95-99.7 rule in a Normal Distribution]


---

### 4. The Probability Density Function (PDF)
The height of the curve at any point $x$ is calculated using this formula:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

* **$\pi$** $\approx 3.14159$
* **$e$** $\approx 2.71828$

---

### 5. Standard Normal Distribution ($Z$-Distribution)
A **Standard Normal Distribution** is a special case where:
* The Mean ($\mu$) = **0**
* The Standard Deviation ($\sigma$) = **1**

We use **$Z$-scores** to "standardize" any normal value into this format, which allows us to compare different datasets:
$$Z = \frac{x - \mu}{\sigma}$$

# üìè The Standard Normal Variate ($Z$)

A **Standard Normal Variate** (often called a **Z-score**) is a specific type of normal random variable that has been "standardized." It represents the number of standard deviations a data point is away from the mean.

---

### 1. Definition
A random variable $X$ is converted into a **Standard Normal Variate ($Z$)** when it is rescaled to have:
* **Mean ($\mu$) = 0**
* **Standard Deviation ($\sigma$) = 1**

Mathematically, we denote this as:  
$$Z \sim N(0, 1)$$



---

### 2. The Standardization Formula (Z-Score)
To transform any value $x$ from a Normal Distribution $N(\mu, \sigma)$ into a Standard Normal Variate $Z$, use the following formula:

$$Z = \frac{x - \mu}{\sigma}$$

* **$x$**: The raw data point.
* **$\mu$**: The mean of the original population.
* **$\sigma$**: The standard deviation of the original population.



---

### 3. Why is $Z$ Important?
Standardizing variables allows us to:
1.  **Compare Apples to Oranges:** You can compare scores from different scales (e.g., comparing an SAT score to an ACT score) by seeing how many standard deviations they are from their respective means.
2.  **Calculate Probabilities Easily:** Instead of calculating complex integrals for every different mean and sigma, we can use a single **Standard Normal Table (Z-Table)**.
3.  **Identify Outliers:** Typically, any $Z$ score greater than $+3$ or less than $-3$ is considered an extreme outlier.

---

### 4. The PDF of the Standard Normal Variate
Because $\mu=0$ and $\sigma=1$, the general Normal PDF simplifies significantly for $Z$:

$$f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$$

---

### üìä Properties of $Z$

| Property | Value |
| :--- | :--- |
| **Mean** | $0$ |
| **Variance** | $1$ |
| **Standard Deviation** | $1$ |
| **Symmetry** | Symmetric around $Z = 0$ |
| **Total Area** | $1.0$ (100% Probability) |

---

### 5. Interpreting $Z$-Scores
* **$Z = 0$**: The value is exactly at the mean.
* **$Z = 1.5$**: The value is 1.5 standard deviations **above** the mean.
* **$Z = -2.0$**: The value is 2.0 standard deviations **below** the mean.

# üìè The Empirical Rule (68-95-99.7 Rule)

The **Empirical Rule**, also known as the **Three-Sigma Rule**, is a statistical guideline used for data that follows a **Normal Distribution**. It describes the percentage of data that falls within specific distances (standard deviations) from the mean.

---

### 1. The Core Rule
In a perfect Normal Distribution, the data is distributed according to these three thresholds:

* **68%** of the data falls within **one** standard deviation ($\mu \pm 1\sigma$).
* **95%** of the data falls within **two** standard deviations ($\mu \pm 2\sigma$).
* **99.7%** of the data falls within **three** standard deviations ($\mu \pm 3\sigma$).



[Image of the Empirical Rule 68-95-99.7 on a normal distribution curve]


---

### 2. Detailed Breakdown
Beyond the three main percentages, we can break the curve down into smaller segments based on symmetry:

* **Between 0 and 1$\sigma$:** ~34.1% of the data.
* **Between 1$\sigma$ and 2$\sigma$:** ~13.6% of the data.
* **Between 2$\sigma$ and 3$\sigma$:** ~2.1% of the data.
* **Beyond 3$\sigma$:** Only ~0.15% of the data (extreme outliers).

---

### 3. When to Use the Empirical Rule
This rule **only** applies when:
1. The data is **symmetrical**.
2. The data is **unimodal** (has one peak).
3. The data follows a **Bell-Shaped (Normal) Distribution**.

> **Note:** If your data is skewed (leaned to one side), you cannot use the Empirical Rule. Instead, you would use **Chebyshev's Theorem**, which applies to all distribution shapes.

---

### 4. Why is it Useful?
* **Outlier Detection:** If a data point has a Z-score greater than 3, the Empirical Rule tells us there is less than a 0.3% chance of it occurring naturally.
* **Quick Probability Estimation:** It allows you to estimate the probability of a value occurring without needing a calculator or a Z-table.
* **Data Validation:** If your data doesn't roughly follow these percentages, it‚Äôs a sign that your data is not Normally Distributed.

---

### üìä Empirical Rule Summary Table

| Distance from Mean | Range Coverage | Probability of falling OUTSIDE |
| :--- | :--- | :--- |
| **$ \mu \pm 1\sigma $** | **68.27%** | ~32% (1 in 3) |
| **$ \mu \pm 2\sigma $** | **95.45%** | ~5% (1 in 20) |
| **$ \mu \pm 3\sigma $** | **99.73%** | ~0.3% (1 in 370) |

---

### üí° Example: Human Height
If the average height of a population is **170 cm** ($\mu$) with a standard deviation of **10 cm** ($\sigma$):
* **68%** of people are between 160 cm and 180 cm.
* **95%** of people are between 150 cm and 190 cm.
* **99.7%** of people are between 140 cm and 200 cm.

# üìè The Empirical Rule (68-95-99.7 Rule)

The **Empirical Rule** describes the percentage of data that falls within specific distances from the mean in a **Normal Distribution**.

### 1. The Core Rule
* **68%** of data falls within **$ \pm 1\sigma $** (one standard deviation).
* **95%** of data falls within **$ \pm 2\sigma $** (two standard deviations).
* **99.7%** of data falls within **$ \pm 3\sigma $** (three standard deviations).

---

### 2. Visualization Code
You can run the following Python code in a Jupyter cell to generate the visualization of this rule.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Setup Data
mu, sigma = 0, 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# 2. Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='black', lw=2)

# Shade 1 Sigma (68%)
x1 = np.linspace(mu - sigma, mu + sigma, 100)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color='skyblue', alpha=0.6, label='68% (1œÉ)')

# Shade 2 Sigma (95%)
x2_l, x2_r = np.linspace(mu-2*sigma, mu-sigma, 100), np.linspace(mu+sigma, mu+2*sigma, 100)
plt.fill_between(x2_l, norm.pdf(x2_l, mu, sigma), color='cornflowerblue', alpha=0.5, label='95% (2œÉ)')
plt.fill_between(x2_r, norm.pdf(x2_r, mu, sigma), color='cornflowerblue', alpha=0.5)

# Shade 3 Sigma (99.7%)
x3_l, x3_r = np.linspace(mu-3*sigma, mu-2*sigma, 100), np.linspace(mu+2*sigma, mu+3*sigma, 100)
plt.fill_between(x3_l, norm.pdf(x3_l, mu, sigma), color='royalblue', alpha=0.4, label='99.7% (3œÉ)')
plt.fill_between(x3_r, norm.pdf(x3_r, mu, sigma), color='royalblue', alpha=0.4)

# 3. Formatting Labels
plt.xticks([-3, -2, -1, 0, 1, 2, 3], ['-3œÉ', '-2œÉ', '-1œÉ', 'Œº', '1œÉ', '2œÉ', '3œÉ'])
plt.title('The Empirical Rule Visualization', fontsize=14)
plt.xlabel('Standard Deviations')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# üìâ Key Properties of the Normal Distribution

The Normal Distribution (or Gaussian Distribution) is defined by its specific mathematical and geometric characteristics. These properties make it the foundation for most statistical inference.

---

### 1. Symmetry and Shape
* **Bell-Shaped:** The graph is a smooth, continuous curve that resembles a bell.
* **Perfect Symmetry:** The distribution is perfectly symmetric around its center. The left half is a mirror image of the right half.
* **Central Tendency:** In a perfect normal distribution, the **Mean, Median, and Mode** are all equal and located at the exact center of the peak.



---

### 2. Parameters: $\mu$ and $\sigma$
The entire shape of the distribution is determined by only two parameters:
1. **Mean ($\mu$):** Shifts the curve left or right (Location).
2. **Standard Deviation ($\sigma$):** Flattens or stretches the curve (Spread).
    * If $\sigma$ is small, the curve is tall and narrow.
    * If $\sigma$ is large, the curve is short and wide.

---

### 3. Total Area and Probability
* **Total Area = 1:** The total area under the density curve is exactly $1$ (or 100%).
* **Probabilistic Nature:** The area under the curve between any two points $a$ and $b$ represents the probability of the variable falling within that range.
* **Calculated as:** $P(a < X < b) = \int_{a}^{b} \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} dx$

---

### 4. The Empirical Rule (68-95-99.7)
This property dictates how data is dispersed relative to the standard deviation:
* **68.27%** of data is within $1\sigma$.
* **95.45%** of data is within $2\sigma$.
* **99.73%** of data is within $3\sigma$.



[Image of the empirical rule 68-95-99.7 on a normal distribution]


---

### 5. Asymptotic Tails
The tails of the normal distribution are **asymptotic** to the x-axis. 
* This means the curve moves closer and closer to the horizontal axis as it moves away from the mean but **never actually touches or crosses it**. 
* Theoretically, this implies that extreme outliers (no matter how far from the mean) have a non-zero probability of occurring.

---

### 6. Statistical Moments
The "shape" metrics for a normal distribution are fixed:
* **Skewness = 0:** Because it is perfectly symmetric, there is no lean to the left or right.
* **Kurtosis = 3:** This is the standard "peakedness" against which all other distributions are measured. (In many software packages, "Excess Kurtosis" is reported as $0$).
