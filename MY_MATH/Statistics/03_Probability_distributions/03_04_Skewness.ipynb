# üìê Skewness: Measuring Asymmetry

**Skewness** is a statistical measure that describes the asymmetry of a probability distribution around its mean. While a Normal Distribution is perfectly symmetrical (zero skew), real-world data often "leans" to one side.

---

### 1. Types of Skewness

#### A. Positive Skew (Right-Skewed)
In a positively skewed distribution, the tail on the right side is longer or fatter.
* **The "Lean":** Data is bunched up on the left (lower values).
* **Mean vs. Median:** The mean is pulled toward the long tail.


[Image of positive skewness distribution]


#### B. Negative Skew (Left-Skewed)
In a negatively skewed distribution, the tail on the left side is longer or fatter.
* **The "Lean":** Data is bunched up on the right (higher values).
* **Mean vs. Median:** The mean is pulled toward the long tail.


[Image of negative skewness distribution]


#### C. Zero Skew (Symmetric)
The distribution is perfectly balanced. The left and right sides are mirror images.
* **Example:** The Normal Distribution.

---

### 2. The Mathematical Formula

Skewness is the **third standardized moment** of a distribution. 

**Population Skewness:**
$$\gamma = E\left[\left(\frac{X - \mu}{\sigma}\right)^3\right]$$

**Sample Skewness:**
$$\text{Skewness} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})^3}{(n-1)s^3}$$

**Whereas:**
* **$x_i$**: Represents each individual value in the dataset.
* **$\bar{x}$**: The **Sample Mean** (the average of your data).
* **$n$**: The total **number of observations** (sample size).
* **$s$**: The **Sample Standard Deviation** (the spread of the data).
* **$\mu$**: The **Population Mean**.
* **$\sigma$**: The **Population Standard Deviation**.
* **$(x_i - \bar{x})^3$**: The **cubed deviation**. Cubing preserves the sign (negative or positive), allowing us to see the direction of the skew.

---

### 3. ‚öîÔ∏è The "VS" Battle: Mean vs. Median vs. Mode

In skewed distributions, the central tendency measures move away from each other in a predictable order.

| Feature | Negative Skew (Left) | Zero Skew (Symmetric) | Positive Skew (Right) |
| :--- | :--- | :--- | :--- |
| **Tail Direction** | Left Side | None (Balanced) | Right Side |
| **Central Tendency** | **Mean < Median < Mode** | **Mean = Median = Mode** | **Mode < Median < Mean** |
| **Average Pull** | Mean is pulled to the **Left** | No pull | Mean is pulled to the **Right** |

[Image comparing mean median and mode positions in positive and negative skewness]

---

### 4. Why Skewness Matters in Data Science

1. **Model Performance:** Many algorithms (like Linear Regression) assume that residuals are normally distributed. Highly skewed features can lead to biased models.
2. **Feature Engineering:** We often use transformations to "fix" skewness:
    * **Log Transform:** $\log(x)$ (Best for Right Skew).
    * **Square Root:** $\sqrt{x}$ (Moderate Right Skew).
    * **Box-Cox:** A generalized power transform.
3. **Outlier Awareness:** High skewness is a signal that your data contains outliers that are dragging the mean away from the "typical" value (the median).

```python

### üêç Python: How to Calculate Skewness

import pandas as pd
import seaborn as sns
from scipy.stats import skew

# Load a sample dataset
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 10, 15] # Example of right-skewed data

# Calculate Skewness
print(f"Skewness: {skew(data)}")

# Visualize
sns.histplot(data, kde=True)
```
