# üìê Skewness: Measuring Asymmetry

**Skewness** is a statistical measure that describes the asymmetry of a probability distribution around its mean. While the Normal Distribution is perfectly symmetrical (zero skew), real-world data often "leans" to one side.

---

### 1. Types of Skewness

#### A. Positive Skew (Right-Skewed)
In a positively skewed distribution, the tail on the right side is longer or fatter than the left side.
* **The "Lean":** Most of the data is bunched up on the left (lower values).
* **Mean vs. Median:** The **Mean** is typically greater than the **Median** because the extreme values in the right tail pull the mean upward.
* **Relationship:** $\text{Mean} > \text{Median} > \text{Mode}$



[Image of positive skewness]


#### B. Negative Skew (Left-Skewed)
In a negatively skewed distribution, the tail on the left side is longer or fatter than the right side.
* **The "Lean":** Most of the data is bunched up on the right (higher values).
* **Mean vs. Median:** The **Mean** is typically less than the **Median** because the extreme low values pull the mean downward.
* **Relationship:** $\text{Mode} > \text{Median} > \text{Mean}$



[Image of negative skewness]


#### C. Zero Skew (Symmetric)
The distribution is perfectly balanced. The left and right sides are mirror images.
* **Example:** The Normal Distribution.
* **Relationship:** $\text{Mean} = \text{Median} = \text{Mode}$

---

### 2. The Mathematical Formula
Skewness is the **third standardized moment** of a distribution. It is calculated as:

$$\gamma = E\left[\left(\frac{X - \mu}{\sigma}\right)^3\right]$$

In simpler terms for a sample:
$$\text{Skewness} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})^3}{(n-1)s^3}$$

* **Positive Result:** Indicates Positive Skew.
* **Negative Result:** Indicates Negative Skew.
* **Zero Result:** Indicates Symmetry.

---

### 3. Comparison Table: Mean, Median, Mode

| Feature | Negative Skew (Left) | Zero Skew (Symmetric) | Positive Skew (Right) |
| :--- | :--- | :--- | :--- |
| **Tail Direction** | Left | None | Right |
| **Peak Position** | Right | Center | Left |
| **Order** | Mean < Median < Mode | Mean = Median = Mode | Mode < Median < Mean |
| **Example** | Age of retirement | Height of adults | Household income |

---

### 4. Why Skewness Matters in Data Science
1.  **Model Assumptions:** Many machine learning models (like Linear Regression) assume that the features follow a Normal Distribution. Highly skewed data can lead to poor model performance.
2.  **Transformation:** If data is highly skewed, we often use mathematical transformations like **Log Transform**, **Square Root**, or **Box-Cox** to make the distribution more "Normal."
3.  **Outlier Detection:** The tail of a skewed distribution often contains the outliers.

```python

### üêç Python: How to Calculate Skewness

import pandas as pd
import seaborn as sns
from scipy.stats import skew

# Load a sample dataset
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 10, 15] # Example of right-skewed data

# Calculate Skewness
print(f"Skewness: {skew(data)}")

# Visualize
sns.histplot(data, kde=True)
```
