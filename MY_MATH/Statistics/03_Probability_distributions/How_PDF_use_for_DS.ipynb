```python
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df = load_iris()
```

```python
df = pd.DataFrame(df.data,columns=df.feature_names,)
```

```python
df
```

```python
iris = load_iris()
df['species'] = iris.target_names[iris.target]
sns.kdeplot(data=df, x='petal length (cm)', hue='species')
sns.ecdfplot(data=df,x=df['petal length (cm)'],hue='species')
```

```python
sns.kdeplot(data=df, x=df['petal width (cm)'], hue='species')
sns.ecdfplot(data=df,x=df['petal width (cm)'],hue='species')
```

```python
sns.kdeplot(data=df, x=df['sepal length (cm)'], hue='species')
sns.ecdfplot(data=df,x=df['sepal length (cm)'],hue='species')
```

```python
sns.kdeplot(data=df, x=df['sepal width (cm)'], hue='species')
sns.ecdfplot(data=df,x=df['sepal width (cm)'],hue='species')
```

# What the above Shows?
- we can see that above the 4 kde distribution only the about 2 distribution (petal length and petal width) show that it help to differenciate the species and we can use them as the feature.
